```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitters=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
chunks=splitters.split_text(text)
```

```python
print(len(chunks))
```

```python
print(chunks[0])
```

```python
def find_relevant_chunk(question):
    for chunk in chunks:
        if question.lower() in chunk.lower():
            return chunk
    return chunks[0]
```

```python
relevant_text=find_relevant_chunk(
    "dataframe"
)
print(relevant_text)
```

```python
question="What is dataframe?"
context=find_relevant_chunk(
    "dataframe"
)
response=model.generate_content(
    f"""
    Answer using this context:
    {context}
    
    Question:
    {question}"""
)   
print(response.text)
```
